```python
def train_perceptron(X, y, learning_rate):
    # w = np.random.random(3)
    w = np.array([5, 1, 1])

    coveraged = False
    num_iter = 0
    while not coveraged:
        coveraged = True

        for X_i, y_i in zip(X, y):
            num_iter += 1
            print("Iteration " + str(num_iter) + ": ", round(w.T @ X_i, 2), round(y_i * w.T @ X_i, 2))
            if y_i * X_i @ w.T <= 0:
                w = w + learning_rate * y_i * X_i
                coveraged = False
            print(w)

    return w
```

```python
X = np.array([[-2, -1],
              [2, -1],
              [1, 1],
              [-1, -1],
              [3, 2]])
X = np.concatenate((np.ones([len(X), 1]), X), axis=1)
y = np.array([-1, 1, 1, -1, 1]).reshape(5, -1)

w = train_perceptron(X, y, 0.4)
print()


# Question b

def sgn(x):
    return 1 if x>0 else -1

x = np.array([1, -5, 3])
wTx = round((w.T @ x), 2)
print("Predicted value = " + str(wTx))
print("Predicted class = " + str(sgn(wTx)))
```
